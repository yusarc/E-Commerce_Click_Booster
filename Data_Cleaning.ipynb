```python
#Missing values of gender %14 percentage, I filled them as 'Unknown'

data['gender'].fillna('Unknown', inplace=True)
```

```python
# Assume that 'has_account' and 'has_price_alert' missing values equal '0'
#Because 'has_account' mean 0.16 and 'has_price_alert' mean 0.24 

data['has_account'].fillna(0, inplace=True)
data['has_price_alert'].fillna(0, inplace=True)
```
